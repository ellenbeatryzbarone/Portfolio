```python
# Instalar bibliotecas necessárias
!pip install pandas matplotlib seaborn plotly

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# -------------------------------------------
# 1. Análise das maiores economias do mundo
# -------------------------------------------

# Dados das maiores economias (exemplo usando dados disponíveis do FMI ou Banco Mundial)
# Aqui, vamos buscar uma fonte atualizada do World Bank
# Para simplicidade, crio um DataFrame fictício com dados atuais
economias = pd.DataFrame({
    'País': ['EUA', 'China', 'Japão', 'Alemanha', 'Reino Unido', 'Indicação', 'França', 'Índia', 'Brasil', 'Itália'],
    'PIB (trilhões USD)': [21.4, 14.3, 5.1, 3.8, 2.7, 1.9, 2.6, 2.9, 1.4, 2.0]
})

# Plotando as maiores economias
plt.figure(figsize=(10,6))
sns.barplot(data=economias.sort_values('PIB (trilhões USD)', ascending=False),
            x='PIB (trilhões USD)', y='País', palette='viridis')
plt.title('Maiores Economias do Mundo (PIB em trilhões USD)')
plt.xlabel('PIB (trilhões USD)')
plt.ylabel('País')
plt.show()

# -------------------------------------------
# 2. Potências com domínio nuclear
# -------------------------------------------

# Dados fictícios ou aproximados de armas nucleares (podem usar fontes como SIPRI)
potencias_nucleares = pd.DataFrame({
    'País': ['EUA', 'Rússia', 'China', 'França', 'Reino Unido', 'Índia', 'Paquistão', 'Israel', 'Coreia do Norte'],
    'Estimativa de Ogivas Nucleares': [5800, 5300, 350, 290, 225, 50, 110, 90, 40]
})

# Gráfico de potência nuclear
plt.figure(figsize=(10,6))
sns.barplot(data=potencias_nucleares.sort_values('Estimativa de Ogivas Nucleares', ascending=False),
            x='Estimativa de Ogivas Nucleares', y='País', palette='magma')
plt.title('Estimativa de Ogivas Nucleares por País')
plt.xlabel('Número de Ogivas Nucleares Estimadas')
plt.ylabel('País')
plt.show()

# Se quiser uma visualização interativa
fig = px.bar(potencias_nucleares, x='País', y='Estimativa de Ogivas Nucleares',
             title='Ogivas Nucleares por País', color='País')
fig.show()
```
